```python
# Calculate percentDeaths and percentRecovered columns
df['percentDeaths'] = (df['totalDeaths'] / df['totalCases']) * 100
df['percentRecovered'] = (df['totalRecovered'] / df['totalCases']) * 100

# Fill NaN values in percentDeaths and percentRecovered with -1
df['percentDeaths'].fillna(-1, inplace=True)
df['percentRecovered'].fillna(-1, inplace=True)

# Rank by percentRecovered, with highest values ranked first
df['rankPercentRecovered'] = df['percentRecovered'].rank(ascending=False, method='first').astype(int)

# Replace non-finite values with NaN in the percentDeaths column
df['percentDeaths'].replace([-1, np.inf, -np.inf], np.nan, inplace=True)

# Calculate rankings without considering NaN values
df['rankPercentDeaths'] = df['percentDeaths'].rank(ascending=True, method='min', na_option='keep').astype(float).astype('Int64')

# Convert rankings to integers
df['rankPercentDeaths'] = df['rankPercentDeaths'].astype('Int64')

# Adjust ranks to -1 for both NaN and -1 values
df.loc[(df['percentRecovered'] == -1), 'rankPercentRecovered'] = -1
df.loc[(df['percentDeaths'] == -1), 'rankPercentDeaths'] = -1

# # Set float formatting to display more decimal places for percentDeaths and percentRecovered
# pd.options.display.float_format = '{:.4f}'.format

pd.set_option('display.max_rows', None)
df
```

```python
country_input = input("Enter a country name: ")

# To handle mismatched cases
country_data = df[df['country'].str.lower() == country_input.lower()]

if not country_data.empty:
    country_name = country_data['country'].values[0]
    percent_deaths = country_data['percentDeaths'].values[0]
    percent_recovered = country_data['percentRecovered'].values[0]
    rank_percent_recovered = country_data['rankPercentRecovered'].values[0]
    rank_percent_deaths = country_data['rankPercentDeaths'].values[0]

    plt.figure(figsize=(10, 6))
    bar_width = 0.2
    bar_positions = range(len(country_data))
    
    plt.barh(bar_positions, country_data['totalCases'], height=bar_width, color='blue', label='Total Cases')
    plt.barh([pos + bar_width for pos in bar_positions], country_data['totalDeaths'], height=bar_width, color='red', label='Total Deaths')
    plt.barh([pos + 2 * bar_width for pos in bar_positions], country_data['totalRecovered'], height=bar_width, color='green', label='Total Recovered')
    
    plt.title(f'{country_name.capitalize()}: Total Cases, Total Deaths, and Total Recovered')
    plt.xlabel('Count')
    plt.ylabel('Country')
    plt.legend()

    # Adding text annotations for additional information
    for pos, val in enumerate(country_data['totalCases']):
        plt.text(val, pos, f'{val:,}', va='center', color='blue')

    for pos, val in enumerate(country_data['totalDeaths']):
        plt.text(val, pos + bar_width, f'{val:,}', va='center', color='red')

    for pos, val in enumerate(country_data['totalRecovered']):
        plt.text(val, pos + 2 * bar_width, f'{val:,}', va='center', color='green')

    info_text = f"percentDeaths: {percent_deaths:.2f}%\npercentRecovered: {percent_recovered:.2f}%\nrankPercentRecovered: {rank_percent_recovered}\nrankPercentDeaths: {rank_percent_deaths}"
    plt.text(0.01, 0.99, info_text, transform=plt.gca().transAxes, va='top', ha='left', fontsize=10, color='black', bbox=dict(facecolor='white', alpha=0.5))

    plt.tight_layout()
    plt.show()
else:
    print(f"No data found for the country: {country_input}")
```
